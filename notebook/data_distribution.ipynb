```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
length = np.load('./data/length.npy')
surp = np.load('./data/surprisingness_96.npy')
norm = np.max(surp)
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=800)

x = np.array([])
y = np.array([])
for i in tqdm(range(len(length[:10000]))):
    x =  np.concatenate((x, np.linspace(0,1,num = length[i])))
    y =  np.concatenate((y, surp[i][:length[i]].squeeze(axis=-1)))
    ax.plot(np.linspace(0,1,num = length[i]),surp[i][:length[i]].squeeze(axis=-1),'k-', linewidth=0.5)

```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=800)
ax.plot(x[:400000],y[:400000],'.',markersize=0.5, linewidth=0.5)
```
